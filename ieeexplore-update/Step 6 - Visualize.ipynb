```python
import pandas as pd
import numpy as np

df = pd.read_csv("IEEE VIS papers 1990-2018 - 2019-Update Preparation(1).csv")
df = df.replace(np.nan, '', regex=True)

df.head(5)
```

```python
toSplit = ['AuthorNames-Deduped', 'AuthorNames','AuthorAffiliation']

toKeep = df.columns[~df.columns.isin(toSplit)]

toKeep.tolist()
```

```python
import numpy as np
from itertools import chain

# return list from series of comma-separated strings
def chainer(s):
    return list(chain.from_iterable(s.str.split(';')))

# calculate lengths of splits
lens = df['AuthorNames-Deduped'].str.split(';').map(len)

# create new dataframe, repeating or chaining as appropriate
res = pd.DataFrame({'DOI': np.repeat(df['DOI'], lens),
                    'Year': np.repeat(df['Year'], lens),
                    'Conference': np.repeat(df['Conference'], lens),
                    'PaperType': np.repeat(df['PaperType'], lens),
                    'Author': chainer(df['AuthorNames-Deduped']),
                    })



toVisualize = res[res["PaperType"] != "M"]
toVisualize = toVisualize[toVisualize["Year"] > 2014]
toVisualize.to_csv("Authors2015-2019.csv")


print(toVisualize.tail())
```

```python
import altair as alt

chart = alt.Chart(toVisualize).mark_bar().encode(
    alt.Y('Author:N', sort='-x'),
    alt.X('count(Author):Q'),
    alt.Color('Conference')
    )

chart

```
