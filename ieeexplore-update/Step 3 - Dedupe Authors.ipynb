```python
import pandas as pd

dblpauthors = pd.read_csv("author-files/dblpAuthors-2022.tsv",keep_default_na=False,sep='\t')


print(dblpauthors.info())
```

```python
currentvispubdataupdate = "vispubdata-current/IEEE VIS papers 1990-2021 - Main dataset-update-in-progress.csv"
vispubdata = pd.read_csv(currentvispubdataupdate)
#vispubdata = pd.read_csv("IEEE VIS papers 1990-2018 - Main dataset.csv")
vispubdata['AuthorNames-Deduped'] = vispubdata['AuthorNames-Deduped'].astype(str)

print(vispubdata.info())
print(vispubdata.index)
```

```python

dblpauthors['DOI-short'] = ""

dblpauthors.loc[dblpauthors['DOI'].str.contains("doi.*org"),'DOI-short'] = dblpauthors['DOI']
dblpauthors['DOI-short'] = dblpauthors['DOI-short'].str.replace("https://doi.*org/","",regex=True)
dblpauthors['DOI-short'] = dblpauthors['DOI-short'].str.replace("http://doi.*org/","",regex=True)


dblpauthors.to_csv("author-files/dblpauthors-DEBUG.csv")
dblpauthors.head(15)
```

```python

dois_not_in_dblp = []

for doi in vispubdata.DOI:
    #find the DOI in the dblpauthors
    
    #print(doi)
    found = dblpauthors['DOI-short'].eq(doi).any() 
    #print(found)
   
    if(found != False):

        dblp_rowindex = dblpauthors.index[dblpauthors['DOI-short'] == doi].tolist()[0]
        #print(dblp_rowindex)
        vispubdata_rowindex = pd.Index(vispubdata.DOI).get_loc(doi)
        #print(vispubdata_rowindex)
        
        vispubdata_deduped_authors = vispubdata.at[vispubdata_rowindex,'AuthorNames-Deduped']

        if vispubdata_deduped_authors != dblpauthors.at[dblp_rowindex,'Authors']:
            dblp_deduped_authors = dblpauthors.at[dblp_rowindex,'Authors']
            print("DOI: " +dblpauthors.at[dblp_rowindex,'DOI'] + " " + vispubdata.at[vispubdata_rowindex,'DOI'])
            print("Existing authors do not match new authors:")
            print(" old: "+str(vispubdata_deduped_authors))
            print(" new: "+str(dblp_deduped_authors))
            vispubdata.at[vispubdata_rowindex,'AuthorNames-Deduped'] = dblp_deduped_authors
        #else:
            #print("Old authors match new authors")
    else:
        dois_not_in_dblp.append(doi)

print(len(dois_not_in_dblp))

```

Create a file for authors and paper DOIs: DedupedAuthors-Papers.csv

```python
newVersion = pd.read_csv(currentvispubdataupdate,dtype=object,keep_default_na=False)
#newVersion = pd.read_csv("IEEE VIS papers 1990-2018 - Main dataset(6).csv",dtype=object,keep_default_na=False)


#create the papers -> author names df
papersAuthorsTemp = pd.DataFrame({'Paper DOI':newVersion['DOI'],'Author Names':newVersion['AuthorNames-Deduped']})
papersAuthors = pd.DataFrame(papersAuthorsTemp['Author Names'].str.split(';').tolist(), index=papersAuthorsTemp['Paper DOI']).stack()
papersAuthors = papersAuthors.reset_index([0, 'Paper DOI'])
papersAuthors.columns = ['Paper DOI', 'Author Names']

papersAuthors.to_csv("author-files/DedupedAuthors-Papers.csv",index=False)

papersAuthors.head()
```

Augment the df above with information on author position

```python
papersAuthorsPosition = papersAuthors
papersAuthorsPosition['PositionNumber'] = -1
papersAuthorsPosition['PositionCode'] = ''
papersAuthorsPosition['Affiliation'] = ''

papersAuthorsPosition.head()
lastDOI = ''
lastPosition = 0
lastindex = -1

#Warning, this assumes that the df was created in sorted order
for index, row in papersAuthorsPosition.iterrows():
    doi = row['Paper DOI']
    position = -1
    affiliation = ''
    
    allAffs = newVersion.loc[newVersion['DOI'] == doi,'AuthorAffiliation'].values[0]
    #print(allAffs)
    
    affList = allAffs.split(";")
    
    if doi != lastDOI:
        #we've reached a new paper
        position = 1
        positionCode = 'F'
        lastDOI = doi
        if len(affList) > 0:
            affiliation = affList[0]
        
        #if we're not at the very first row
        if lastindex != -1:
            #we need to update the positionCode of the last author
            papersAuthorsPosition.loc[lastindex,'PositionCode'] = 'L'
            #if it was a single author, we set them to 'F'
            if papersAuthorsPosition.loc[lastindex,'PositionNumber'] == 1:
                papersAuthorsPosition.loc[lastindex,'PositionCode'] = 'F'

    elif doi == lastDOI:
        #we're still at the same paper
        if len(affList) > lastPosition:
            affiliation = affList[lastPosition]
        position = lastPosition + 1
        positionCode = 'M'
        
        
    papersAuthorsPosition.loc[index,'PositionNumber'] = position
    papersAuthorsPosition.loc[index,'PositionCode'] = positionCode
    papersAuthorsPosition.loc[index,'Affiliation'] = affiliation
    lastindex = index
    lastPosition = position
        
        
papersAuthorsPosition.to_csv("author-files/DedupedAuthors-Papers-Position-Affiliation.csv",index=False)

papersAuthorsPosition.head(10)
```

### Create the new affiliation file for cleaning

```python
#load the previously cleaned affilation file
cleanedAffiliations = pd.read_csv("author-files/authors-affiliations-cleaned-June-2020.csv",dtype=object,keep_default_na=False)
#print(cleanedAffiliations.head())

mergedDF = papersAuthorsPosition.merge(cleanedAffiliations,how='left',left_on=['Paper DOI','Author Names'],right_on=['DOI','AuthorNames-Deduped'])
mergedDF.drop(['Year','DOI','AuthorNames-Deduped'],axis=1,inplace=True)
mergedDF.columns = ['DOI','AuthorName-Deduped','PositionNumber','PositionCode','OriginalAffiliation','CleanedAffiliation','Country']


mergedDF.to_csv('author-files/authors-affiliations-2022-toclean-withVASTconf.csv')

mergedDF.head()
```
