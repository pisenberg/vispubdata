# Compare Aminer & Vispubdata
This notebook runs some code to check the difference between titles not found in Aminer (Natkamon's code) and DBLP

```python
import pandas as pd
import editdistance

dblpauthors = pd.read_csv("dblpAuthors.csv",keep_default_na=False)

print(dblpauthors.info())

vispubdata = pd.read_csv("IEEE VIS papers 1990-2018 - Main dataset.csv")
print(vispubdata.info())
```

```python
titles_not_found = vispubdata[vispubdata['AminerCitationCount_02-2019'].isnull()]
titles_not_found = titles_not_found[titles_not_found['PaperType'] != "M"]
print(titles_not_found.info())
```

```python

vpd_titles = []
dblp_bestmatch = []
levenshtein = []

for t in titles_not_found.Title:
    mindistance = 999999
    bestmatch = ""
    for dblp_title in dblpauthors.PublTitle:
        distance = editdistance.eval(t,dblp_title)
        if distance < mindistance:
            mindistance = distance
            bestmatch = dblp_title
    print("Best match: " + t + " - " + bestmatch)
    vpd_titles.append(t)
    dblp_bestmatch.append(bestmatch)
    levenshtein.append(mindistance)
    
pd.DataFrame(vpd_titles,dblp_bestmatch,distance).to_csv('Aminer-VPD-DBLP-BestMatches.csv',index=False)
```

```python
print(len(levenshtein))
print(len(vpd_titles))
print(len(dblp_bestmatch))

#{'Authors':deduped_unique_authors}

pd.DataFrame({'Vispubdata title':vpd_titles,'DBLP title best match':dblp_bestmatch,'Levenshtein distance':levenshtein}).to_csv('Aminer-VPD-DBLP-BestMatches.csv',index=False)
```
